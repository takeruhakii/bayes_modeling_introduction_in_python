# 時系列分析と状態空間モデルの基礎
一般化線形モデルなどの多くのモデルがデータが独立同時分布(i.i.d)に従うと仮定してモデル化される。しかし、この仮定は非常に強い仮定であり、従わないデータも多い。時系列データはその一つであり、現在のデータが過去のデータの実現値に依存して生成される。このようなデータに対するモデリングでは、潜在変数がマルコフ連鎖を構成し、潜在変数から観測値が得られるという状態空間モデルを考えることが多い。潜在変数が離散値を取る場合を隠れマルコフモデル(HMM)、潜在変数が連続値でガウス分布に従う場合を動的線形モデル(DLM)と呼ぶ。また、観測値がガウス分布以外の分布に従う時、動的一般化線形モデル(DGLM)と呼ぶ。

状態空間モデルでは、過去の観測値を元に現在の状態を更新し、将来の状態を得て、将来の状態から観測値の予測値を得ることができる。また、観測値が欠損している場合、欠損期間の状態を推定することで欠損値の補完をすることができる。

## 状態空間モデル
状態空間モデルでは、時系列データのデータ生成過程をモデル化することができる。ある時点$t$の潜在変数(状態)を$\alpha_t$, 観測値を$y_t$とすると、一般化状態空間モデルは次のように書ける。
$$
\alpha_t \sim f(\alpha_t|\alpha_{t-1}) \quad (1)\\
y_t \sim g(y_t|\alpha_t)\quad (2)
$$
(1)は状態の変化を表すので、状態モデル,状態方程式と呼ばれ、(2)は観測値が得られるプロセスを表すので観測モデル,観測方程式と呼ばれる。

## 動的線形モデル 動的一般化線形モデル
動的線形モデルは、観測変数と状態変数がガウス分布に従うと仮定したモデルであり、状態変数は過去の状態変数の線形結合で表される線形モデル、観測値は状態変数の線形接合で表される線形モデルでモデル化される。

$\xi_t$を過程誤差, $\epsilon_t$を観測誤差とした時、動的線形モデルは次のように表される。
$$
\boldsymbol{\alpha_t} = \boldsymbol{T_t}\boldsymbol{\alpha_{t-1}} + \boldsymbol{R_t}\boldsymbol{\xi_t} \quad
\boldsymbol{\xi_t} \sim Normal(0, \boldsymbol{Q_t}) \\
\boldsymbol{y_t} = \boldsymbol{Z_t}\boldsymbol{\alpha_t} + \boldsymbol{\epsilon_t} \quad
\boldsymbol{\epsilon_t} \sim Normal(0,\boldsymbol{H_t})
$$
$\boldsymbol{T_t}, \boldsymbol{R_t}, \boldsymbol{Z_t}$はモデルの構造を決める行列である。

さらに、線形モデルから一般化線形モデルへの拡張のように、動的線形モデルも観測値の従う分布と活性化関数(リンク関数)を変えることで動的一般化線形モデルに拡張することができる

## 状態空間モデルの推定
状態空間モデルはMCMCを用いて事後分布を推定することもできるが、隠れマルコフモデルではEMアルゴリズム、動的線形モデルではカルマンフィルタを用いて最尤推定値を推定することもできる。
